# 02 — Analyse des Features URL
Visualisation des 18 features structurelles

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.features.url_features import extract_features_batch, FEATURE_NAMES

df = pd.read_parquet('../data/processed/train.parquet')
X = extract_features_batch(df['url'])
feat_df = pd.DataFrame(X, columns=FEATURE_NAMES)
feat_df['label'] = df['label'].values
feat_df.head()
```

```python
# Corrélation features vs label
corr = feat_df.corr()['label'].drop('label').sort_values(key=abs, ascending=False)
corr.plot(kind='barh', figsize=(8, 6), color='steelblue')
plt.title('Corrélation features vs label phishing')
plt.axvline(0, color='black', lw=0.5)
plt.tight_layout()
plt.show()
```

```python
# Distribution de quelques features clés
key_features = ['url_length', 'domain_entropy', 'digit_ratio', 'subdomain_depth']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, feat in zip(axes.flatten(), key_features):
    for label, color in [(0, 'green'), (1, 'red')]:
        feat_df[feat_df.label==label][feat].hist(
            ax=ax, alpha=0.6, color=color, bins=40, label=['Légit','Phish'][label]
        )
    ax.set_title(feat)
    ax.legend()

plt.tight_layout()
plt.show()
```
