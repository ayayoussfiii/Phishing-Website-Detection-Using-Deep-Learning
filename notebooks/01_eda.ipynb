# 01 — Exploration du Dataset
Analyse des URLs PhishTank + Alexa

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_parquet('../data/processed/train.parquet')
print(train_df.shape)
train_df.head(10)
```

```python
# Distribution des labels
train_df['label'].value_counts().plot(kind='bar')
plt.title('Distribution phishing (1) vs légitimes (0)')
plt.show()
```

```python
# Longueur des URLs
train_df['url_len'] = train_df['url'].str.len()
train_df.groupby('label')['url_len'].hist(alpha=0.7, bins=50)
plt.title('Distribution longueur URL')
plt.legend(['Légitimes', 'Phishing'])
plt.show()
```

```python
# Exemples d'URLs phishing vs légitimes
print('=== URLs PHISHING ===')
for url in train_df[train_df.label==1]['url'].head(5):
    print(' -', url)

print('\n=== URLs LÉGITIMES ===')
for url in train_df[train_df.label==0]['url'].head(5):
    print(' -', url)
```
