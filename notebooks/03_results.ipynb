# 03 — Entraînement & Résultats
Comparaison des 3 modèles

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, roc_auc_score

from src.features.feature_pipeline import FeaturePipeline

test_df = pd.read_parquet('../data/processed/test.parquet')
pipeline = FeaturePipeline.load('../models/pipeline')
X_seq, X_feat, y_test = pipeline.transform(test_df)
print(f'Test set: {len(y_test):,} samples')
```

```python
# Charger et évaluer le modèle hybride
hybrid = tf.keras.models.load_model('../models/hybrid_best.keras')
y_prob = hybrid.predict([X_seq, X_feat], verbose=0).flatten()
y_pred = (y_prob > 0.5).astype(int)

print('=== Modèle Hybride ===')
print(classification_report(y_test, y_pred, target_names=['Légitimes', 'Phishing']))
print(f'AUC-ROC : {roc_auc_score(y_test, y_prob):.4f}')
```
